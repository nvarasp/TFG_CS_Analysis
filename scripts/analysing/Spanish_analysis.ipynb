## GETTING CLEAN NAMES

```python
SP_initiatives["clean_name"]=""
for index, row in SP_initiatives.iterrows():
    #print("NAME NORMAL: ", row["name"])
    if (len(row["name"].split('('))>1):
        if(len(row["name"].split('(')[0].split('–'))>1):
            SP_initiatives.loc[index, "clean_name"]=remove_text_marks(row["name"].split('(')[0].split('–')[0])
        else:
            SP_initiatives.loc[index, "clean_name"]=remove_text_marks(row["name"].split('(')[0])
        
    elif (len(row["name"].split(':'))>1):
        SP_initiatives.loc[index, "clean_name"]=remove_text_marks(row["name"].split(':')[0])
    elif (len(row["name"].split('–'))>1):
        SP_initiatives.loc[index, "clean_name"]=remove_text_marks(row["name"].split('–')[0])
    elif (len(row["name"].split('-'))>1):
        SP_initiatives.loc[index, "clean_name"]=remove_text_marks(row["name"].split('-')[0])
    elif (len(row["name"].split('.'))>1):
        SP_initiatives.loc[index, "clean_name"]=remove_text_marks(row["name"].split('.')[0])
    else:
        SP_initiatives.loc[index, "clean_name"]=row["name"]
    #print("CLEANED: ", row["clean_name"])
```

```python
tfidfs = []
SP_initiatives["aim_resultwords"]=np.empty((len(SP_initiatives), 0)).tolist()
document_size=len(SP_initiatives)
# calculate tf
for index, row in SP_initiatives.iterrows():
    tfidf = TFIDF(row["aim_clean"])
    tfidf.tf()
    tfidfs.append(tfidf)
# calculate idf

all_text = [tfidf.text for tfidf in tfidfs]

for i, tfidf in enumerate(tfidfs):
    tfidf.idf(all_text, document_size)
    tfidf.tf_idf()
    result_df = tfidf.text_df.sort_values('tfidf', ascending=False).head(5)  # sort and take top 5
    result_words = []
    for index, row in result_df.iterrows():
        result_words.append((row.word, row.tfidf))
    SP_initiatives.loc[i, "aim_resultwords"]=result_words
```

```python
tfidfs = []
SP_initiatives["description_resultwords"]=np.empty((len(SP_initiatives), 0)).tolist()
document_size=len(SP_initiatives)
# calculate tf
for index, row in SP_initiatives.iterrows():
    tfidf = TFIDF(row["description_clean"])
    tfidf.tf()
    tfidfs.append(tfidf)
# calculate idf

all_text = [tfidf.text for tfidf in tfidfs]

for i, tfidf in enumerate(tfidfs):
    tfidf.idf(all_text, document_size)
    tfidf.tf_idf()
    result_df = tfidf.text_df.sort_values('tfidf', ascending=False).head(5)  # sort and take top 5
    result_words = []
    for index, row in result_df.iterrows():
        result_words.append((row.word, row.tfidf))
    SP_initiatives.loc[i, "description_resultwords"]=result_words
```

```python
humanities_social=['Ciencias Sociales','Cultura y Arqueología', 'Culturas Indígenas', 'Ciencias Políticas','Geografía','Educación']
physical_sciences=['Espacio y Astronomía','Geología y Ciencias de la Tierra', 'Física', 'Ciencias Químicas']
life_sciences=['Ecología y Medioambiente', 'Biodiversidad', 'Biología', 'Agua', 'Marino y Terrestre', 'Océanos', 
               'Naturaleza y Aire Libre', 'Animales', 'Pájaros', 'Seguimiento de Especies a largo plazo', 
               'Clima y Meteorología',  'Insectos y Polinizadores', 'Genética', 'Biogeografía']
applied_sciences=['Informática y Ciencias de la Computación', 'Medicina y Salud','Sonido',
                  'Ciencias de la Agricultura y Veterinaria','Transporte','Ciencia de los Alimentos','Gestión de Recursos Naturales']
```

```python
topics_list=['humanities_social', 'physical_sciences', 'life_sciences', 'applied_sciences']
count_humanities=0
count_physical=0
count_life=0
count_applied=0

topics_SP=SP_initiatives[['id', 'name', 'keywords']].copy()
for element in topics_list:
    topics_SP[element]=False
    
for index, row in SP_initiatives.iterrows():
    if any(x in row['keywords'] for x in humanities_social):
        topics_SP.loc[index,["humanities_social"]]=True
        count_humanities+=1
    if any(x in row['keywords'] for x in physical_sciences):
        topics_SP.loc[index,["physical_sciences"]]=True
        count_physical+=1
    if any(x in row['keywords'] for x in life_sciences):
        topics_SP.loc[index,["life_sciences"]]=True
        count_life+=1
    if any(x in row['keywords'] for x in applied_sciences):
        topics_SP.loc[index,["applied_sciences"]]=True
        count_applied+=1
```
