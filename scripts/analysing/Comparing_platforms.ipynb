## COMPARING PROJECTS THAT ARE SHOWN IN MORE THAN ONE PLATFORM

Print project names that are repeated in two of the platforms

```python
#See the projects that are posted in both European and Spanish platforms
repeated_projects_names_EU_SP_europename=[]
repeated_projects_names_EU_SP_spanishname=[]

for project_name in EU_initiatives.clean_name:
    for project_name_2 in SP_initiatives.clean_name:
        if project_name in project_name_2 and project_name_2!="Observado" :#since observado is present in more than one project name, and it is not the correct name of the original project, that is why it's skipped
            repeated_projects_names_EU_SP_spanishname.append(project_name_2)
            #print(project_name)
            #print(project_name_2)
repeated_projects_names_EU_SP_spanishname
```

```python
#See the projects that are posted in both European and BCN platforms
repeated_projects_names_EU_BCN_europename=[]
repeated_projects_names_EU_BCN_bcnname=[]

for project_name in EU_initiatives.clean_name:
    for project_name_2 in BCN_initiatives.name:
        if project_name in project_name_2:
            repeated_projects_names_EU_BCN_europename.append(project_name)
            repeated_projects_names_EU_BCN_bcnname.append(project_name_2)
            #print(project_name)
            #print(project_name_2)
        
repeated_projects_names_EU_BCN_europename
```

```python
#See the projects that are posted in both Spanish and BCN platforms
repeated_projects_names_SP_BCN_spanishname=[]
repeated_projects_names_SP_BCN_bcnname=[]
for project_name in SP_initiatives.clean_name:
    for project_name_2 in BCN_initiatives.name:
        if project_name in project_name_2 and project_name!="Observado" :
            repeated_projects_names_SP_BCN_spanishname.append(project_name)
            repeated_projects_names_SP_BCN_bcnname.append(project_name_2)
            #print(project_name)
            #print(project_name_2)
repeated_projects_names_SP_BCN_bcnname
```

```python
#See the projects that are posted in the three platforms
repeated_projects_names_3_europeanname=[]
repeated_projects_names_3_spanishname=[]
repeated_projects_names_3_bcnname=[]
        
for project_name in EU_initiatives.clean_name:
    for project_name_2 in SP_initiatives.clean_name:
        for project_name_3 in BCN_initiatives.name:
            if project_name_2 in project_name_3 and project_name in project_name_3 and project_name_2!="Observado":
                repeated_projects_names_3_europeanname.append(project_name)
                repeated_projects_names_3_spanishname.append(project_name_2)
                repeated_projects_names_3_bcnname.append(project_name_3)
repeated_projects_names_3_bcnname
```

## JOINING ALL PROJECTS TO GET A GLOBAL OVERVIEW

```python
MIX_projects = EU_initiatives[["name", "start_date", "end_date", "country", "description_resultwords", "needed_list", "topics_list", "status_", "topic_social", "topic_physical", "topic_life", "topic_applied", "duration_years", "education_in_text", "education_in_keywords_topic", "clean_name"]]
MIX_projects.insert(len(MIX_projects.columns), "origin_platform", "EU-citizen.science")
```

```python
#ADD SPANISH PROJECTS
for index, row in SP_initiatives.iterrows():
    if row["clean_name"] not in repeated_projects_names_EU_SP_spanishname:
        new_row = {"name": row["name"], "start_date": row["start_date"], "end_date": row["end_date"],
                   "country": "ES", "description_resultwords": row["description_resultwords"], 
                   "needed_list": row["needed_list"], "topics_list": row["keywords_list"], "status_": None, "topic_social": row["topic_social"], 
                   "topic_physical": row["topic_physical"], "topic_life": row["topic_life"], "topic_applied": row["topic_applied"], "duration_years": row["duration_years"],  
                   "education_in_text": row["education_in_text"], "education_in_keywords_topic": (row["education_in_keywords"] |row["education_in_impact"]), "clean_name": row["clean_name"],
                  "origin_platform": "Observatorio CC"}
        MIX_projects=MIX_projects.append(new_row, ignore_index=True)
```

```python
#ADD BARCELONIAN PROJECTS
for index, row in BCN_initiatives.iterrows():
    if row["name"] not in repeated_projects_names_SP_BCN_bcnname and row["name"] not in repeated_projects_names_EU_BCN_bcnname:
        if row["phone"]==True:
            needed="phone"
        else:
            needed=None
            
        if row["topic"]=="social":
            social=True
            physical=False
            life=False
            applied=False
        elif row["topic"]=="salud":
            social=False
            physical=False
            life=False
            applied=True
        elif row["topic"]=="ambiental" or  row["topic"]=="ambiental, salud":
            social=False
            physical=False
            life=True
            applied=False
        else:
            social=False
            physical=False
            life=False
            applied=False
            
            
        new_row = {"name": row["name"], "start_date": None, "end_date": None,
                   "country": "ES", "description_resultwords": row["description_resultwords"], 
                   "needed_list": needed, "topics_list": row["topic"], "status_": row["status"], "topic_social": social, 
                   "topic_physical": physical, "topic_life": life, "topic_applied": applied, "duration_years": None,  
                   "education_in_text": None, "education_in_keywords_topic": None, "clean_name": row["name"],
                  "origin_platform": "Oficina de CC"}
        MIX_projects=MIX_projects.append(new_row, ignore_index=True)
```

```python
#EXPORT TO EXCEL
with pd.ExcelWriter('MIX_projects.xlsx', options={'strings_to_urls': False}) as writer:
        MIX_projects.to_excel(writer, 'MIX_projects')
MIX_projects[:5]
```
